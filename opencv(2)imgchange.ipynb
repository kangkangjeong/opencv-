```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import sys 
import os
```

```python
img=cv2.imread('./puppy.bmp',cv2.IMREAD_REDUCED_COLOR_2)
h,w=img.shape[:2]
img_resize=cv2.resize(img,(h//2,w//2))
if img is None:
    print('image read failed')
    sys.exit()

cv2.namedWindow('image',cv2.WINDOW_NORMAL)
cv2.imshow('image',img)


while True:
    key=cv2.waitKey()
    if key==27 or key==ord('q'):  
        break
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
img_BGR=cv2.imread('./puppy.bmp',cv2.IMREAD_REDUCED_COLOR_2)
img_RGB=cv2.cvtColor(img_BGR,cv2.COLOR_BGR2RGB)
img_GRAY=cv2.cvtColor(img_BGR,cv2.COLOR_BGR2GRAY)

if img_BGR is None:
    print('image load failed')
    sys.exit()

fig ,ax= plt.subplots(1,3,figsize=(12,6))

ax[0].imshow(img_BGR),ax[0].set_title('img_BGR'),ax[0].axis('off')
ax[1].imshow(img_RGB),ax[1].set_title('img_BR'),ax[1].axis('off')
ax[2].imshow(img_GRAY,cmap='gray'),ax[2].set_title('img_GR'),ax[2].axis('off')

plt.show()
```

```python
img_paths=[]
img_list=os.listdir('./my_images/')
for i in img_list:
    
    img_path='./my_images/'+i
    img_paths.append(img_path)
cv2.namedWindow('image',cv2.WINDOW_NORMAL)
cv2.setWindowProperty('image',cv2.WND_PROP_FULLSCREEN,
                      cv2.WINDOW_FULLSCREEN)
idx=0
while True:
    img=cv2.imread(img_paths[idx])   
        
    if img is None:
        print('image read failed')
        continue
    
    cv2.imshow('image',img)
    if cv2.waitKey(3000)==27:
        break
    idx +=1
    
    if idx >= len(img_paths):
        idx=0

cv2.destroyAllWindows()
```

```python
src = cv2.imread('fig/cow.png')
dst = cv2.imread('fig/field.bmp')
h,w = src.shape[:2]
dst_resize=cv2.resize(dst,(w,h))

src_gray=cv2.cvtColor(src,cv2.COLOR_BGR2GRAY)
_,mask=cv2.threshold(src_gray,240,255,cv2.THRESH_BINARY_INV)

cv2.copyTo(src,mask,dst_resize)
cv2.imshow('src',src)
cv2.imshow('mask',mask)
cv2.imshow('dst_resize',dst_resize)
cv2.imwrite('./fig/cowingreen.jpg',dst_resize)
cv2.waitKey()
cv2.destroyAllWindows()
```
