```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
im=cv2.imread('./puppy.bmp',cv2.IMREAD_COLOR)
print(im.shape)
im_rgb=cv2.cvtColor(im,cv2.COLOR_BGR2RGB)
im_gray=cv2.cvtColor(im,cv2.COLOR_BGR2GRAY)
if im is None:
    print('error')
    sys.exit()
    
cv2.namedWindow('image',cv2.WINDOW_AUTOSIZE)
cv2.imshow('image',im)

while True:
    key=cv2.waitKey()
    if key==27 or key==ord('q'):  #아스키코드로 인식 27은 esc 하지만 ord를 쓰면 아스키코드로 바꿔준다.
        break
cv2.destroyAllWindows()
```

```python
fig, ax = plt.subplots(1, 3, figsize=(15, 5))
ax[0].imshow(im), ax[0].set_title('BGR'), ax[0].axis('off')
ax[1].imshow(im_rgb), ax[1].set_title('RGB'), ax[1].axis('off')
ax[2].imshow(im_gray, cmap='gray'), ax[2].set_title('GRAY'), ax[2].axis('off')

plt.show()
```

```python
import os
import cv2
```

```python
img_list=os.listdir('./my_images/')
```

```python
img_paths=[]

for i in img_list:
    
    img_path='./my_images/'+i
    img_paths.append(img_path)
cv2.namedWindow('image',cv2.WINDOW_NORMAL)
cv2.setWindowProperty('image',cv2.WND_PROP_FULLSCREEN,
                      cv2.WINDOW_FULLSCREEN)
idx=0
while True:
    img=cv2.imread(img_paths[idx])   
        
    if img is None:
        print('image read failed')
        continue
    
    cv2.imshow('image',img)
    if cv2.waitKey(3000)==27:
        break
    idx +=1
    
    if idx >= len(img_paths):
        idx=0

cv2.destroyAllWindows()
```
