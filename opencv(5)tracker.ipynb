```python
import cv2
import sys

cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print('Failed to open video capture')
    sys.exit()

# # Define the codec and create a VideoWriter object
# width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
# height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
# fps = int(cap.get(cv2.CAP_PROP_FPS))

# fourcc = cv2.VideoWriter_fourcc(*'XVID')
f_flages=False
e_flages=False
while True:

    ret, frame = cap.read()

    if not ret:
        print('Failed to capture frame')
        break
    if f_flages:
        frame=cv2.flip(frame,1)
        
    if e_flages:
        frame = cv2.Canny(frame,50,100)
    cv2.imshow('frame',frame)
    key=cv2.waitKey(20)
    
    if key==27:
        break
    elif key==ord('f'):
        f_flages=not f_flages
    elif key==ord('e'):
        e_flages=not e_flages

cap.release()
cv2.destroyAllWindows()
```

```python
# cv2.setMouseCallback(windowName, onMouse, param = None) -> None
    # windowName: 마우스이벤트를 수행할 창 이름
    # onMouse: 마우스 이벤트 콜벡함수
    # param: 콜백함수에 전달할 데이터

# onMouse(event, x, y, flags, param) -> None
# event: 마우스 이벤트 종류 e.g., cv2.EVENT_LBUTTONDOWN
# x, y : 창을 기준으로 이벤트 발생좌표
# flags: 이벤트시 발생 상태 e.g., "ctrl"
# param: cv2.setMouseCallback()함수에서 설정한 데이터

# 마우스 event와 flags reference
# docs/opencv.org/master -> MouseEventTypes cv, MouseEventFlags cv
```

```python
import numpy as np
import cv2

def call_mouse(event,x,y,flags,param):
    #마우스클릭하면 xy좌표를 받는다
    if event==cv2.EVENT_LBUTTONDOWN:
        print('left button down: \n',x,y)
    elif event==cv2.EVENT_LBUTTONUP:
        print('left button up: \n',x,y)
    elif event==cv2.EVENT_MOUSEMOVE:
        #누르면실행되게
        if flags == cv2.EVENT_FLAG_LBUTTON:
          print(x,y)


img= np.ones((600,600,3),np.uint8)*255

cv2.namedWindow('image')
cv2.setMouseCallback('image',call_mouse)
cv2.imshow('image',img)

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
#하얀색 배경에 마우스로 그림그리고 r로 리프레시 s로 저장
import numpy as np
import cv2

def call_mouse(event,x,y,flags,param):
    global oldx, oldy
    if event==cv2.EVENT_LBUTTONDOWN:
        oldx,oldy=x,y

    elif event == cv2.EVENT_MOUSEMOVE:
        if flags == cv2.EVENT_FLAG_LBUTTON:
            cv2.line(img, (oldx, oldy), (x, y), (0, 0, 255), 5, cv2.LINE_AA)
            cv2.imshow('image', img)
            oldx, oldy = x, y


img= np.ones((600,600,3),np.uint8)*255

cv2.namedWindow('image')
cv2.setMouseCallback('image',call_mouse)

while True:
    cv2.imshow('image', img)
    key = cv2.waitKey()
    if key == ord('r'):
        img = np.ones((600, 600, 3), np.uint8) * 255
    elif key==ord('s'):
        cv2.imwrite('image.png',img)
    elif key == 27:
        break
cv2.imshow('image',img)
cv2.destroyAllWindows()
```
